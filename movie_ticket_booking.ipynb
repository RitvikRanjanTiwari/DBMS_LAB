```python
import mysql.connector
cnx=mysql.connector.connect(host='localhost',port=3306,user='root',password='',database='lab_assignment1')
cursor=cnx.cursor(buffered=True)
cursor.execute('SHOW TABLES')
for x in cursor:
    print(x)
```

```python
query = """
        create table city(
            city_id varchar(8) not null primary key,
            name varchar(15)
        )
        """
cursor.execute(query)
print('city table created!')
```

```python
query = """
        create table theatre(
        theatre_id varchar(8) not null primary key,
        name varchar(15),
        address varchar(30),
        city_id varchar(8),
        foreign key (city_id) references city(city_id)
        )
        """
cursor.execute(query)
print('theatre table created')
```

```python
query = """  create table movie(
            movie_id varchar(8) not null primary key,
            name varchar(15),
            genre varchar(10),
            duration int,
            release_date date
        )
        """
cursor.execute(query)
print('movie table created')
```

```python
query = """
        create table screens(
            screen_id varchar(8) not null,
            name varchar(15),
            price int,
            total_seating int,
            seats_filled int,
            theatre_id varchar(8),
            foreign key(theatre_id) references theatre(theatre_id),
            movie_id varchar(8),
            foreign key (movie_id) references movie(movie_id),
            show_timings time not null,
            primary key (screen_id,show_timings)
        )
        """
cursor.execute(query)
print('screens table created!')

```

```python
query = """
        create table bookings(
            booking_id varchar(8) not null primary key,
            user_name varchar(20),
            mobile_number varchar(10),
            email varchar(15),
            booked_seats int,
            booking_date datetime,
            screen_id varchar(8),
            show_timings time,
            foreign key (screen_id,show_timings) 
            REFERENCES screens(screen_id,show_timings)
        )
        """

cursor.execute(query)
print('booking table created!')
```

```python
from datetime import datetime
from IPython.display import clear_output

class MTBS:
    def __init__(self,cursor):
        self.cursor = cursor

    def show_cities(self):
        query = "select * from city"
        self.cursor.execute(query)
        cities = self.cursor.fetchall()
        return cities
    
    def show_movies(self,city_id):
        query = """
                select movie_id,name from movie M join
                        (
                         select distinct movie_id as ID 
                         from screens join theatre
                         on screens.theatre_id = theatre.theatre_id
                         where theatre.city_id = %s
                        ) T
                on M.movie_id = T.ID
                """
        self.cursor.execute(query,[city_id])
        movies = self.cursor.fetchall()
        return movies
    
    def show_theatre(self,city_id,movie_id):
        query = """
                select distinct theatre.theatre_id, theatre.name 
                from screens join theatre
                on theatre.theatre_id = screens.theatre_id
                where (city_id,movie_id) = (%s,%s)
                """
        self.cursor.execute(query,[city_id,movie_id])
        theatres = self.cursor.fetchall()
        return theatres
    
    def show_screen_timing(self,theatre_id,movie_id):
        query = """
                select distinct screen_id,name,
                show_timings,total_seating,seats_filled,price 
                from screens
                where (theatre_id,movie_id) = (%s,%s) 
                and total_seating <> seats_filled
                """
        self.cursor.execute(query,[theatre_id,movie_id])
        screens = self.cursor.fetchall()
        return screens
    
    def show_available_seats(self,screen_id,show_timings):
        query = """
                select * from screens
                where (screen_id, show_timings) = (%s,%s)
                """
        self.cursor.execute(query,[screen_id,show_timings])
        availability = self.cursor.fetchall()
        return availability
    
    def generate_booking_id(self):
        query = 'select * from bookings'
        self.cursor.execute(query)
        res = self.cursor.fetchall()
        return 'MTBS'+str(len(res)+1).rjust(4, '0')

    def book_tickets(self):
        
        clear_output()
        print('\n=>\n')
        
    # select a city 
        print("\nList of available cities : ")   
        cities = self.show_cities()
        for i in range(len(cities)):
            print(i+1,' ',cities[i][1])
        selected_city = int(input('Enter your option : '))
        city_name = cities[selected_city-1][1]
        
        clear_output()
        print('\n=> {} =>\n'.format(city_name))
        
    # select a movie 
        print('\nMovies available in {} : '.format(city_name))
        movies = self.show_movies(cities[selected_city-1][0])
        for i in range(len(movies)):
            print(i+1,' ',movies[i][1])
        
        selected_movie = int(input('Enter your option : '))
        movie_name = movies[selected_movie-1][1]
        
        clear_output()
        print('\n=> {} => {} =>\n'.format(city_name,movie_name))
        
    # select a theatre
        print('\n{} movie is available in : '.format(movie_name))
        theatres = self.show_theatre(
            cities[selected_city-1][0],
            movies[selected_movie-1][0]
        )
        for i in range(len(theatres)):
            print(i+1,' ',theatres[i][1])
        
        selected_theatre = int(input('Enter your option : '))
        theatre_name = theatres[selected_theatre-1][1]
        
        clear_output()
        print('\n=> {} => {} => {} =>\n'.format(
            city_name,movie_name,theatre_name))
        
    # select screen and show timing
        print('\n{} movie show timings in {} : '.format(
            movie_name,theatre_name))
        screen_timings = self.show_screen_timing(
            theatres[selected_theatre-1][0],
            movies[selected_movie-1][0])
        for i in range(len(screen_timings)):
            print(i+1,' ',
                  screen_timings[i][1],' : ',
                  screen_timings[i][2],
                  ' : {} seats available : price [ {} ]'.format(
                      screen_timings[i][3]-screen_timings[i][4],
                      screen_timings[i][5]))
        
        selected_screen_timings = int(input('Enter your option : '))
        screen_name = screen_timings[i][1]
        timings = screen_timings[i][2]
        
        
        clear_output()
        print('\n=> {} => {} => {} => {} => {} =>\n'.format(
            city_name,movie_name,theatre_name,screen_name,timings))
        
    # number of seats for booking
        print("\nselect number of seats : ")
        seat_count = int(input())

        clear_output()
        print('\n=> {} => {} => {} => {} => {} => {} seats\n'.format(
            city_name,movie_name,theatre_name,screen_name,timings,
            seat_count))
        
        print("\nPersonal Details\n")
        user_name = input("Name : ")
        mobile_number = input("Mobile Number : ")
        email = input("Email : ")

        clear_output()
        
    # Review Booking
        print("\n\n Review Booking\n")
        print('Movie : ',movies[selected_movie-1][1])
        print('Theatre : ',theatres[selected_theatre-1][1])
        print('Screen : ',screen_timings[selected_screen_timings-1][1])
        print('Show Timings : ',
              screen_timings[selected_screen_timings-1][2])
        print('Seats selected : ',seat_count)
        print('Amount to be paid : ',
              seat_count*int(screen_timings[selected_screen_timings-1][5]))

        print('\nConfirm your booking : ')
        confirm = input("\nconfirm / cancel ? : ")
        
        clear_output()
        
        if confirm == 'confirm':
        
        # New Booking tuple
            booking_id = self.generate_booking_id()
            date_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            query = """
                    insert into bookings 
                    values(%s,%s,%s,%s,%s,%s,%s,%s)
                    """
            self.cursor.execute(query,
                                [booking_id,
                                 user_name,
                                 mobile_number,
                                 email,seat_count,
                                 date_time,
                                 screen_timings[selected_screen_timings-1][0],
                                 screen_timings[selected_screen_timings-1][2]])
            cnx.commit()

        # Reduce the seats booked
            query = """update screens set seats_filled = %s 
            where screen_id= %s and show_timings = %s"""
            cursor.execute(query,
                [screen_timings[selected_screen_timings-1][4]+seat_count ,
                 screen_timings[selected_screen_timings-1][0],
                 screen_timings[selected_screen_timings-1][2]])
            cnx.commit()
        
        # printing booking confirmation
            print('_______________________________________________')
            print('\n\nTickets booked !\n\n')
            print('Booking ID \t: ',booking_id)
            print('Name \t\t: ',user_name)
            print('Email \t\t: ',email)
            print('Mobile Number \t: ',mobile_number)
            print('_______________________________________________')
            print('\nMovie \t\t: ',movies[selected_movie-1][1])
            print('Theatre \t: ',theatres[selected_theatre-1][1])
            print('Screen \t\t: ',
                  screen_timings[selected_screen_timings-1][1])
            print('Show Timings \t: ',
                  screen_timings[selected_screen_timings-1][2])
            print('Seats selected \t: ',seat_count)
            print('_______________________________________________')
            
        else:
            print('Booking Canceled!')
```
